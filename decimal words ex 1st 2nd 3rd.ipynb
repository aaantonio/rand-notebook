```python
def notwin(place):
    for x in range(101):
        if x != place:
            if str(x)[-1] == '1' and len(str(x)) < 2:
                print(str(x) + 'st')
            elif len(str(x)) > 1 and str(x)[-2] == '1':
                print(str(x) + 'th')
            elif str(x)[-1] == '1':
                print(str(x) + 'st')
            elif str(x)[-1] == '2':
                print(str(x) + 'nd')
            elif str(x)[-1] == '3':
                print(str(x) + 'rd')
            else:
                print(str(x) + 'th')
```

```python
notwin(4)
```
